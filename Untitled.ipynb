```python
import pandas as pd
import numpy as np
import folium
from folium.plugins import MarkerCluster
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

# 1. Load Data

```python
historico = pd.read_csv("historico_delitos_cdmx.txt", sep="|")
delitos = pd.read_csv("catalogo_delitos.txt", sep="|")
```

```python
delitos.head()
```

```python
historico.head()
```

```python
# Just to see the joined datasets

delitos = delitos.rename(columns={'ID': 'tipo_delito'})
merged = pd.merge(historico, delitos, on='tipo_delito')
merged.head()
```

```python
merged['delito'].value_counts()
```

In the above dataframe we can see the joined dataset. Let's handle both dataframe separately to visualizate some insights.

```python
crimes = merged.groupby(['AÑO','delito']).size().unstack()
crimes.head()
```

```python
merged.groupby(['AÑO', 'delito']).mean()
```

```python
merged.groupby(['MES', 'delito']).mean()
```

```python
merged.groupby(['MES','delito']).size().unstack()
```

```python
plotiar = merged.groupby(['AÑO','delito']).size().unstack()
```

```python
plotiar.plot(kind='bar',figsize=(10,10))
```

```python
plotiar.plot(kind='bar', stacked=True, figsize=(10,10))
```

# Geographic Visualization

```python
MEXICO_COORDINATES = (19.432608, -99.133209)

#Please select the year and the kind of crime
Year=2014
Kind_Crime=7
historic_2014 = (merged[(merged['AÑO']==Year) & (merged['tipo_delito']==Kind_Crime)])

historic_map = folium.Map(location=[19.432608, -99.133209],tiles='stamentoner', zoom_start=12)

marker_cluster = MarkerCluster().add_to(historic_map)

for index,row in historic_2014.iterrows():
    folium.Marker(location=[row['CY'],row['CX']],popup=str(row['delito']),icon=None,).add_to(marker_cluster)

print("===============Map showing crime(number): "+str(Kind_Crime)+" in year "+str(Year)+"=====================")
historic_map
    
```

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

# 1. Load Data

```python
historico = pd.read_csv("historico_delitos_cdmx.txt", sep="|")
delitos = pd.read_csv("catalogo_delitos.txt", sep="|")
```

```python
delitos.head()
```

```python
historico.head()
```

```python
# Just to see the joined datasets

delitos = delitos.rename(columns={'ID': 'tipo_delito'})
merged = pd.merge(historico, delitos, on='tipo_delito')
merged.head()
```

```python
merged['delito'].value_counts()
```

In the above dataframe we can see the joined dataset. Let's handle both dataframe separately to visualizate some insights.

```python
crimes = merged.groupby(['AÑO','delito']).size().unstack()
crimes.head()
```

```python
merged.groupby(['AÑO', 'delito']).mean()
```

```python
merged.groupby(['MES', 'delito']).mean()
```

```python
merged.groupby(['MES','delito']).size().unstack()
```

```python
plotiar = merged.groupby(['AÑO','delito']).size().unstack()
```

```python
plotiar.plot(kind='bar',figsize=(10,10))
```

```python
plotiar.plot(kind='bar', stacked=True, figsize=(10,10))
```
